```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("edgeflower.jpg")  
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
blur = cv2.GaussianBlur(gray, (5, 5), 0)
```

```python
edges_main = cv2.Canny(blur, 100, 200) 
```

```python
params = [(50,150), (100,200), (150,250)]
edges_list = []

for (t1, t2) in params:
    e = cv2.Canny(blur, t1, t2)
    edges_list.append((t1, t2, e))
```

```python
plt.figure(figsize=(15,10))

# Original Image
plt.subplot(2,3,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis("off")

# Grayscale Image
plt.subplot(2,3,2)
plt.title("Grayscale")
plt.imshow(gray, cmap="gray")
plt.axis("off")

# Blurred Image
plt.subplot(2,3,3)
plt.title("Gaussian Blur")
plt.imshow(blur, cmap="gray")
plt.axis("off")

# Canny Edges (100,200)
plt.subplot(2,3,4)
plt.title("Canny Edges (100,200)")
plt.imshow(edges_main, cmap="gray")
plt.axis("off")

# Parameter Comparison
plt.subplot(2,3,5)
plt.title("Canny Comparison")
plt.axis("off")

plt.subplot(2,3,5)
plt.title("Canny (50,150)")
plt.imshow(edges_list[0][2], cmap="gray")
plt.axis("off")

plt.subplot(2,3,6)
plt.title("Canny (150,250)")
plt.imshow(edges_list[2][2], cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```
